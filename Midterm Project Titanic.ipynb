```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# load training set
training_set = pd.read_csv(os.getcwd() + '/Data/train.csv')
training_set.head()
```

```python
#load testing set
testing_set = pd.read_csv(os.getcwd() + '/Data/test.csv')
testing_set.head()

```

```python
# testing_set.insert(1, 'Survived', 'default value')

# load missing data from gender_submission and append to testing set
# undo this step
# testing_set_survival = pd.read_csv(os.getcwd() + '/Data/gender_submission.csv')
# testing_set['Survived'] = testing_set_survival['Survived']
# testing_set
```

```python
#combine training and testing set SUCCESS
#delete cabin column because too many missing value
data = training_set.append(testing_set, ignore_index=True)
data.shape
del data['Cabin']
data
```

```python
#describion/notation of data
data.describe().T
```

```python
#visualize data
data.hist(figsize=(20,15))
plt.show()
```

```python

data.plot(kind="scatter",x="Age",y="Fare",alpha=0.1)
```

```python
#scatter plot
from pandas.plotting import scatter_matrix
attributes = ["Survived", "Fare", "Age","Sex"]
scatter_matrix(data[attributes], figsize=(12, 8))
plt.show()
```

```python
#age and fare tell us who payed the most fare and how old was that person.
agefare = data[['Age' ,'Fare']]
plt.plot(agefare['Age'],agefare['Fare'], 'b.')
```

```python

# data['Sex'] = data['Sex'].apply({1:'male', 0:'female'}.get)
# data['Sex'].replace({0:'female',1:'male'},inplace=True)

#filling the missing values in Age and Fare with its mean SUCCEED
data['Age'].fillna((data['Age'].mean()), inplace = True)
data['Fare'].fillna((data['Fare'].mean()), inplace = True)

#change sex column to display male as 1 and female as 0 SUCCEED
data.Sex[data.Sex == 'male'] = 1
data.Sex[data.Sex == 'female'] = 0
data
```

```python
# delete rows of missing embarked value
# replace missing string in Embarked with NaN then remove row with NaN Embarked value
data['Embarked'].replace('', np.nan, inplace=True)
data.dropna(subset=['Embarked'], inplace=True)
data.info()
```

```python
# attempt to create 2 new column 1 contain value true if a person is a Dr and the other for Master.
```

```python
# make 3 variable to categorize Embarked values: Embark_S, Embark_C, and Embark_Q SUCCEED
data_cleaned = pd.get_dummies(data, prefix=['Embarked'], columns=['Embarked'])
data_cleaned
#data_cleaned is cleared of missing string but variables is not scaled
```

```python
# Separate Name and Ticket column from the data
# separate combined data into training and testing set, train set has notnull survival column while test set has otherwise.
data_num = data_cleaned.drop(['Name', 'Ticket'], axis=1)
data_num_col = data_num.columns
data_num_train = data_num.loc[data_num['Survived'].notnull(), data_num_col]
data_num_test = data_num.loc[data_num['Survived'].isnull(), data_num_col]
data_num_test.head()
```

```python
data_num_train.head()
```

```python
# This step is moved do after scaling  
# to avoid scaling the Survival column we are going to save it in a different variable
# data_num_train["Survived"] = data_num_train['Survived'].astype(int)
# data_num_train_survived = data_num_train['Survived']
# data_num_train_survived
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

data_num_train = scaler.fit_transform(data_num_train)
data_num_train = pd.DataFrame(data_num_train,
                            columns=data_num_col)
data_num_test = scaler.fit_transform(data_num_test)
data_num_test = pd.DataFrame(data_num_test,
                            columns = data_num_col)

data_num_test.head()
```

```python
# append back the data in 0 and 1 form after scaling data
# data_num_train['Survived'] = data_num_train_survived
# to avoid scaling the Survival column we are going to save it in a different variable
data_num_train["Survived"] = data_num_train['Survived'].astype(int)
# data_num_train_survived = data_num_train['Survived']

#turned out 2 rows of survived values of training set was null so we remove them
data_num_train.dropna(subset=['Survived'], inplace=True)
data_num_train
```

```python
# Select and train models
target = data_num_train['Survived']
target
```

```python
#split data into train_set and vali_set from the data_num_train

data_num_train_final = data_num_train.drop(['Survived'], axis=1)
from sklearn.model_selection import train_test_split
T_train, T_test, V_train, V_test = train_test_split(data_num_train_final, target, test_size=0.2)
```

```python
V_train.shape
```

```python
V_train[:20]
```

# Use SGDClassifier 

```python
# Use SGDClassifier 
from sklearn.linear_model import SGDClassifier
sgd = SGDClassifier()
sgd.fit(T_train, V_train)
```

```python
#make a prediction on train set using sgd
sgd_train_predict = sgd.predict(T_train)
sgd_train_predict[:10]
```

```python
# Calculate the train accuracy
from sklearn.metrics import accuracy_score
train_accuracy = accuracy_score(V_train, sgd_train_predict)
print('train_accuracy:', train_accuracy)
```

```python
#cross value score
from sklearn.model_selection import cross_val_score
accuracy_scores = cross_val_score(sgd, data_num_train_final, target, cv=5,
                                 scoring='accuracy')
print(accuracy_scores)
```

```python
# confusion matrix of train set
from sklearn.metrics import confusion_matrix
conf_mat = confusion_matrix(V_train, sgd_train_predict)
print(conf_mat)
```

```python
# precision, recall and f1 scores
from sklearn.metrics import precision_score, recall_score, f1_score
precision = precision_score(V_train, sgd_train_predict)
recall = recall_score(V_train, sgd_train_predict)
f1 = f1_score(V_train, sgd_train_predict)
print(precision, recall, f1)
```

```python
# ROC (Receiver operation characteristic) curve
from sklearn.metrics import roc_curve
test_scores = sgd.decision_function(T_train)
tpr, npr, thresholds = roc_curve(V_train, sgd_train_predict)

# plot tpr vs. npr.
plt.plot(tpr, npr, 'b-')
```

```python
# Area under curve score
from sklearn.metrics import roc_auc_score
score = roc_auc_score(V_train, sgd_train_predict)
print("AUC score:", score)
```

# Use LOGISTIC REGRESSION

```python
# 
#You are passing floats to a classifier which expects categorical values as the target vector. 
#If you convert it to int it will be accepted as input (although it will be questionable if that's the right way to do it).

from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
log.fit(T_train, V_train)
```

```python
log_train_predict = log.predict(T_train)
log_train_predict[:10]
```

```python
# Calculate the test accuracy
from sklearn.metrics import accuracy_score
train_accuracy = accuracy_score(V_train, log_train_predict)
print('train_accuracy:', train_accuracy)
```

```python
#cross value score
from sklearn.model_selection import cross_val_score
accuracy_scores = cross_val_score(log, data_num_train_final, target, cv=5,
                                 scoring='accuracy')
print(accuracy_scores)
```

```python
# confusion matrix
from sklearn.metrics import confusion_matrix
conf_mat = confusion_matrix(V_train, log_train_predict)
print(conf_mat)
```

```python
# precision, recall and f1 scores
from sklearn.metrics import precision_score, recall_score, f1_score
precision = precision_score(V_train, log_train_predict)
recall = recall_score(V_train, log_train_predict)
f1 = f1_score(V_train, log_train_predict)
print(precision, recall, f1)
```

```python
# ROC (Receiver operation characteristic) curve
from sklearn.metrics import roc_curve
train_scores = log.decision_function(T_train)
tpr, npr, thresholds = roc_curve(V_train, train_scores)

# plot tpr vs. npr.
plt.plot(tpr, npr, 'b-')
```

```python
# Area under curve score
from sklearn.metrics import roc_auc_score
score = roc_auc_score(V_train, train_scores)
print("AUC score:", score)
```
